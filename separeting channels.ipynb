```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Read the image using OpenCV
input_image_path = "1st_png/mfcc_png/blues/mfcc_2.png"
image = cv2.imread(input_image_path)
image_arrray=np.array(image)

print(image_arrray)

# Split the image into its RGB channels
blue_channel, green_channel, red_channel = cv2.split(image)

# Plot histograms for each channel
plt.figure(figsize=(15, 5))

# Blue channel histogram
plt.subplot(1, 3, 1)
plt.hist(blue_channel.ravel(), bins=256, color='blue', alpha=0.5) # to view complete 2D use ravel from cv2 for 1D use flatten to convert 2D array to 1D and then visualize
plt.title('Blue Channel')
plt.xlim([0, 255])

# Green channel histogram
plt.subplot(1, 3, 2)
plt.hist(green_channel.ravel(), bins=256, color='green', alpha=0.5)
plt.title('Green Channel')
plt.xlim([0, 255])

# Red channel histogram
plt.subplot(1, 3, 3)
plt.hist(red_channel.ravel(), bins=256, color='red', alpha=0.5)
plt.title('Red Channel')
plt.xlim([0, 255])

plt.show()
```
